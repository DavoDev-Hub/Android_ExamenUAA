Cada ejercicio tiene un valor de 1.25 puntos.
Instrucciones: escribe tu propuesta de código para resolver cada uno de los siguientes problemas.
- Solicita el nombre, edad y altura de una persona. Identifica si es mayor o menor de edad. Y muestra un mensaje con el siguiente: “Joaquin, altura 1.7 metros, dado que tu edad es de 16 años eres menor de edad”.

- Solicita una calificación numérica del 0 al 100, y muestra su equivalente en letra:
    - a. 90 a 100
    - b. 80 a 89
    - c. 70 a 79
    - d. 60 a 69
    - e. 50 a 59
    - f. Menor a 50

- Dada una lista de 5 números y muestra el número mayor, menor y el promedio.

- Solicita una contraseña y verifica que cumpla las siguientes condiciones (muestra las condiciones que no cumple):
    - La contraseña debe contener un mínimo de 8 caracteres.
    - La contraseña debe tener al menos una letra minúscula.
    - La contraseña debe tener al menos una letra mayúscula.
    - La contraseña debe tener al menos un número.
    - La contraseña debe tener al menos un carácter especial (¡#$%&=).

- Crea una función que reciba el precio de un producto, un porcentaje de descuento (por defecto 10) y un booleano si es miembro (por defecto false). Si el cliente es miembro, se aplica un 5% adicional de descuento. La función debe retornar el precio final. Llama a la función al menos en tres formas diferentes.

- Solicita un número del 1 al 10 y muestra la tabla de multiplicar. Si el número es par, muestra la tabla en orden ascendente; si es impar, muestra la tabla en orden descendente.

- Crea una clase que contenga el nombre, el precio y la cantidad de un producto. Agrega un método para calcular el total (precio y cantidad). Crea una lista de 4 productos diferentes, muestra el total de cada producto y el total final.

- Crea una clase que contenga el id, el nombre y la calificación final del estudiante. Incluye una función que muestre si el estudiante aprueba (>= 6.5) o no. Crea 3 instancias y muestra si aprueban o no.
