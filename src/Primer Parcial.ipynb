- Solicita una contraseña y verifica que cumpla las siguientes condiciones (muestra las condiciones que no cumple):
    - La contraseña debe contener un mínimo de 8 caracteres.
    - La contraseña debe tener al menos una letra minúscula.
    - La contraseña debe tener al menos una letra mayúscula.

    - La contraseña debe tener al menos un carácter especial (¡#$%&=).

```kotlin
println("Introduce una contraseña:")
val contrasena = readln()

val tieneMinimo8 = contrasena.length >= 8
val tieneMinuscula = contrasena.any { it.isLowerCase() }
val tieneMayuscula = contrasena.any { it.isUpperCase() }
val tieneNumero = contrasena.any { it.isDigit() }
val tieneEspecial = contrasena.any { it in "¡#\$%&=*" }

if (!tieneMinimo8) {
    println("La contraseña debe tener al menos 8 caracteres")
}
if (!tieneMinuscula) {
    println("La contraseña debe tener al menos una letra minúscula")
}
if (!tieneMayuscula) {
    println("La contraseña debe tener al menos una letra mayúscula")
}
if (!tieneNumero) {
    println("La contraseña debe tener al menos un número")
}
if (!tieneEspecial) {
    println("La contraseña debe tener al menos un carácter especial (¡#\$%&=)")
}

if (tieneMinimo8 && tieneMinuscula && tieneMayuscula && tieneNumero && tieneEspecial) {
    println("La contraseña es válida")
}
```

- Crea una función que reciba el precio de un producto, un porcentaje de descuento (por defecto 10) y un booleano si es miembro (por defecto false). Si el cliente es miembro, se aplica un 5% adicional de descuento. La función debe retornar el precio final. Llama a la función al menos en tres formas diferentes.

```kotlin
fun calcularPrecio(precio: Double, descuento: Int = 10, esMiembro: Boolean = false): Double {
    val descuentoFinal = if (esMiembro) descuento + 5 else descuento
    val precioFinal = precio - (precio * descuentoFinal / 100)
    return precioFinal
}

val precio1 = calcularPrecio(100.0)
println("Precio sin descuento personalizado: $precio1")

val precio2 = calcularPrecio(100.0, 20)
println("Precio con 20% de descuento: $precio2")

val precio3 = calcularPrecio(100.0, 20, true)
println("Precio con 20% de descuento y siendo miembro: $precio3")

val precio4 = calcularPrecio(100.0, esMiembro = true)
println("Precio con descuento por defecto siendo miembro: $precio4")
```

- Solicita un número del 1 al 10 y muestra la tabla de multiplicar. Si el número es par, muestra la tabla en orden ascendente; si es impar, muestra la tabla en orden descendente.

```kotlin
println("Introduce un numero del 1 al 10:")
val numero = readln().toIntOrNull()

if (numero == null || numero !in 1..10) {
    println("Introduce un número válido entre 1 y 10")
} else {
    if (numero % 2 == 0) {
        println("Tabla del $numero (ascendente):")
        for (i in 1..10) {
            println("$numero x $i = ${numero * i}")
        }
    } else {
        println("Tabla del $numero (descendente):")
        for (i in 10 downTo 1) {
            println("$numero x $i = ${numero * i}")
        }
    }
}
```

- Crea una clase que contenga el nombre, el precio y la cantidad de un producto. Agrega un método para calcular el total (precio y cantidad). Crea una lista de 4 productos diferentes, muestra el total de cada producto y el total final.

```kotlin
class Producto(
    val nombre: String,
    val precio: Double,
    val cantidad: Int
) {
    fun calcularTotal(): Double {
        return precio * cantidad
    }
}

val productos = listOf(
    Producto("Laptop", 15000.0, 1),
    Producto("Mouse", 250.0, 2),
    Producto("Teclado", 500.0, 1),
    Producto("Monitor", 3000.0, 2)
)

var totalFinal = 0.0

for (producto in productos) {
    val totalProducto = producto.calcularTotal()
    println("${producto.nombre}: $totalProducto")
    totalFinal += totalProducto
}

println("Total final: $totalFinal")
```

- Crea una clase que contenga el id, el nombre y la calificación final del estudiante. Incluye una función que muestre si el estudiante aprueba (>= 6.5) o no. Crea 3 instancias y muestra si aprueban o no.

```kotlin
class Estudiante(
    val id: Int,
    val nombre: String,
    val calificacionFinal: Double
) {
    fun verificarAprobacion(): String {
        return if (calificacionFinal >= 6.5) {
            "$id $nombre aprueba"
        } else {
            "$id $nombre no aprueba"
        }
    }
}

val estudiante1 = Estudiante(280622, "Ana", 8.0)
val estudiante2 = Estudiante(298721, "Luis", 6.0)
val estudiante3 = Estudiante(334612, "Carlos", 7.5)

println(estudiante1.verificarAprobacion())
println(estudiante2.verificarAprobacion())
println(estudiante3.verificarAprobacion())
```
